```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Prosper Loans dataset
prosper_loans = pd.read_csv(r"C:/Users/M_Osaid/Downloads/prosperLoanData.csv")

# Basic Data Exploration
# Check the first five rows of the dataset
prosper_loans.head()

# Check the shape of the dataset
prosper_loans.shape

# Check the data types of the columns
prosper_loans.dtypes

# Check the summary statistics of the numerical columns
prosper_loans.describe()

# Check the unique values of categorical columns
prosper_loans.select_dtypes(include=['object']).nunique()

# Check for missing values
prosper_loans.isnull().sum()

# Data Visualization
# Histogram of loan amount
plt.hist(prosper_loans['LoanOriginalAmount'])
plt.title('Distribution of Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()

# Boxplot of loan amount by loan status
sns.boxplot(x='LoanStatus', y='LoanOriginalAmount', data=prosper_loans)
plt.title('Loan Amount by Loan Status')
plt.show()

# Heatmap of correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(prosper_loans.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Scatterplot of borrower rate vs. credit score range
plt.scatter(x='CreditScoreRangeLower', y='BorrowerRate', data=prosper_loans)
plt.title('Borrower Rate vs. Credit Score Range')
plt.xlabel('Credit Score Range Lower')
plt.ylabel('Borrower Rate')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Prosper Loans dataset
prosper_loans = pd.read_csv(r"C:/Users/M_Osaid/Downloads/prosperLoanData.csv")

# Basic Data Exploration
# Check the summary statistics of the 'LoanStatus' column
prosper_loans['LoanStatus'].describe()

# Check the frequency distribution of 'LoanStatus' column
prosper_loans['LoanStatus'].value_counts()

# Data Visualization
# Barplot of loan status count
sns.countplot(x='LoanStatus', data=prosper_loans)
plt.title('Loan Status Count')
plt.xticks(rotation=90)
plt.show()

# Pie chart of loan status count
plt.pie(prosper_loans['LoanStatus'].value_counts(), labels=prosper_loans['LoanStatus'].unique(), autopct='%1.1f%%')
plt.title('Loan Status Proportion')
plt.show()

# Boxplot of loan original amount by loan status
sns.boxplot(x='LoanStatus', y='LoanOriginalAmount', data=prosper_loans)
plt.title('Loan Amount by Loan Status')
plt.xticks(rotation=90)
plt.show()

# Barplot of borrower rate by loan status
sns.barplot(x='LoanStatus', y='BorrowerRate', data=prosper_loans)
plt.title('Borrower Rate by Loan Status')
plt.xticks(rotation=90)
plt.show()

# Violinplot of loan term by loan status
sns.violinplot(x='LoanStatus', y='Term', data=prosper_loans)
plt.title('Loan Term by Loan Status')
plt.xticks(rotation=90)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the Prosper Loans dataset
#prosper_loans = pd.read_csv('ProsperLoanData.csv')

# Data Preparation
# Create a new column for loan status category
prosper_loans['LoanStatusCategory'] = np.where(prosper_loans['LoanStatus'].isin(['Completed', 'FinalPaymentInProgress']), 'Paid Off', 'Defaulted or Current')

# Drop irrelevant columns
prosper_loans = prosper_loans.drop(columns=['ListingKey', 'ListingNumber', 'LoanKey', 'LoanNumber', 'LoanStatus'])

# Data Cleaning
# Drop rows with missing values
prosper_loans = prosper_loans.dropna()

# Remove outliers from the 'BorrowerRate' and 'LoanOriginalAmount' columns
prosper_loans = prosper_loans[prosper_loans['BorrowerRate'] <= 0.5]
prosper_loans = prosper_loans[prosper_loans['LoanOriginalAmount'] <= 35000]

# Data Encoding
# Convert categorical variables to numerical using one-hot encoding
prosper_loans = pd.get_dummies(prosper_loans, columns=['EmploymentStatus', 'LoanStatusCategory'])

# Convert the Prosper Rating (Alpha) column to a numerical scale
prosper_loans['ProsperRating (numeric)'] = prosper_loans['ProsperRating (Alpha)'].map({'HR': 1, 'E': 2, 'D': 3, 'C': 4, 'B': 5, 'A': 6, 'AA': 7})

# Drop the original Prosper Rating (Alpha) column
prosper_loans = prosper_loans.drop(columns=['ProsperRating (Alpha)'])
```
